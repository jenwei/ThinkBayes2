# CDF

```python
import thinkbayes2
import thinkplot 
```

```python
dice6=thinkbayes2.Pmf(range(1,7))
three=dice6+dice6+dice6
```

```python
thinkplot.Hist(three)
```

```python
cdf=three.MakeCdf()
```

```python
thinkplot.Cdf(cdf)
```

```python
cdf = three.Max(6)
thinkplot.Cdf(cdf)
```

#  THE DICE PROBLEM

```python
from dice import Dice
d={4:2,6:6,8:3,10:2,12:4,20:4}
pmf = Dice(d)
```

```python
thinkplot.Hist(pmf)
```

```python
pmf.Update(6) #data = 6 (roll a 6)
```

```python
thinkplot.Hist(pmf)
```

```python
cdf=pmf.MakeCdf()
thinkplot.Cdf(cdf)
```

If we roll a 6 again

```python
pmf.Update(6)
```

```python
thinkplot.Hist(pmf)
```

For the third roll, we roll a 5

```python
pmf.Update(5)
```

```python
thinkplot.Hist(pmf)
```

For the fourth roll, we roll an 8

```python
pmf.Update(8)
```

```python
thinkplot.Hist(pmf)
```

```python
cdf=pmf.MakeCdf()
thinkplot.Cdf(cdf)
```
